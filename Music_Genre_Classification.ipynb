```python
import numpy as np

class BroadLearning:
    def __init__(self, num_mapped_features, num_nonlinear_mappings):
        self.num_mapped_features = num_mapped_features
        self.num_nonlinear_mappings = num_nonlinear_mappings
        self.W_e = np.random.randn(num_mapped_features, num_mapped_features)
        self.beta_e = np.random.randn(num_mapped_features)
        self.W_h = np.random.randn(num_mapped_features, num_nonlinear_mappings)
        self.beta_h = np.random.randn(num_nonlinear_mappings)
        self.W = None

    def linear_projection(self, X):
        Z = []
        for i in range(self.num_mapped_features):
            Z_i = np.dot(X, self.W_e[i]) + self.beta_e[i]
            Z.append(Z_i)
        return np.array(Z)

    def nonlinear_mapping(self, Z):
        H = []
        for j in range(self.num_nonlinear_mappings):
            H_j = np.tanh(np.dot(Z, self.W_h[:, j]) + self.beta_h[j])
            H.append(H_j)
        return np.array(H)

    def train(self, X, Y):
        Z = self.linear_projection(X)
        H = self.nonlinear_mapping(Z)
        combined_features = np.concatenate((Z, H), axis=1)
        self.W = np.linalg.pinv(combined_features) @ Y

    def predict(self, X_cap):
        Z_cap = self.linear_projection(X_cap)
        H_cap = self.nonlinear_mapping(Z_cap)
        combined_features_cap = np.concatenate((Z_cap, H_cap), axis=1)
        Y_cap = combined_features_cap @ self.W
        return Y_cap
```

```python
# Set num_mapped_features based on the number of features extracted by the RCNN model
num_mapped_features = X_train.shape[1]  # Assuming X_train_features is your extracted features

# Set num_nonlinear_mappings
num_nonlinear_mappings = 2  # Example value, you can experiment with different values

# Initialize Broad Learning model
broad_learning_model = BroadLearning(num_mapped_features=num_mapped_features, num_nonlinear_mappings=num_nonlinear_mappings)

# Train the Broad Learning model
broad_learning_model.train(X_train, y_train)

# Predict tags for test data
Y_pred = broad_learning_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# Compute evaluation metrics
threshold = 0.5
Y_pred_binary = (y_pred > threshold).astype(int)
accuracy = accuracy_score(y_test, Y_pred_binary)
precision = precision_score(y_test, Y_pred_binary, average='binary')
recall = recall_score(y_test, Y_pred_binary, average='binary')
f1 = f1_score(y_test, Y_pred_binary, average='binary')

# Print evaluation metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
